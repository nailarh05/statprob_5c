```python
plt.figure(figsize=(20, 18))

plt.subplot(3, 1, 1)
plt.boxplot(data['Curah Hujan (mm)'])
plt.title('Boxplot Curah Hujan (mm)')
plt.ylabel('Curah Hujan (mm)')

plt.subplot(3, 1, 2)
plt.boxplot(data['Penjualan A (pcs)'])
plt.title('Boxplot Penjualan A (pcs)')
plt.ylabel('Penjualan A (pcs)')

plt.subplot(3, 1, 3)
plt.boxplot(data['Penjualan B (pcs)'])
plt.title('Boxplot Penjualan B (pcs)')
plt.ylabel('Penjualan B (pcs)')

plt.tight_layout()
plt.show()

# Deteksi Outlier
def detect_outliers(column):

    Q1 = np.percentile(column, 25)
    Q3 = np.percentile(column, 75)
    # Rentang Interquartile (IQR)
    IQR = Q3 - Q1
    # Batas bawah dan atas untuk mendeteksi outlier
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Deteksi outlier
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers

# Mendeteksi outlier untuk setiap variabel
outliers_hujan = detect_outliers(data['Curah Hujan (mm)'])
outliers_a = detect_outliers(data['Penjualan A (pcs)'])
outliers_b = detect_outliers(data['Penjualan B (pcs)'])

print(f"Outliers Curah Hujan (mm): {outliers_hujan}")
print(f"\nOutliers Penjualan A (pcs): {outliers_a}")
print(f"\nOutliers Penjualan B (pcs): {outliers_b}")
```
